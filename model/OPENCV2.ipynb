```python
import cv2
import matplotlib.pyplot as plt
```

```python
coco_file = "ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
forzen_model = "frozen_inference_graph.pb"
Labels = "coco.names"
```

```python
model = cv2.dnn_DetectionModel(forzen_model,coco_file)
```

```python
model
```

```python
classlabel = []
with open(Labels,'rt') as fpt:
    classlabel = fpt.read().rstrip('\n').split('\n')
    
```

```python
len(classlabel)
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

# Reading image

```python
img = cv2.imread("1.jpg")
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
Classindex, confidece, bbox = model.detect(img, confThreshold = 0.5)
```

```python
print(Classindex)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for Classind, conf, boxes in zip(Classindex.flatten(), confidece.flatten(), bbox):
    cv2.rectangle(img, boxes, (225, 0, 0), 4)
    
    # Check if Classind is within the valid range
    if 0 <= Classind - 1 < len(classlabel):
        class_name = classlabel[Classind - 1]
        cv2.putText(img, class_name, (boxes[0] + 10, boxes[1] + 40), font, fontScale=font_scale, color=(0, 255, 0), thickness=3)
    else:
        print(f"Invalid Classind: {Classind}")
```
